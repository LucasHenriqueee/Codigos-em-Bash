# AMOSTRAGEM


# Bibliotecas

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

# Funções


```python
def gerador_da_funcao_senoide(x):
    y = np.sin(2*np.pi*2*x)
    return y, x

def amostrador(fs, n):
    N = np.arange(n)
    t = N / fs
    return t

def plot(sinal_processado, t, fs):

  fig = make_subplots(
  rows=3, cols=1,
  shared_xaxes=False,
  vertical_spacing=0.1)


  # Traço com hastes (todas, inclusive com y=0)
  lines_x = []
  lines_y = []


  # Marcadores (somente onde y ≠ 0)
  marker_x = []
  marker_y = []

  for i in range(len(t)):
      if not np.isclose(sinal_processado[i], 0):
        marker_x.append(i)
        marker_y.append(sinal_processado[i])

  fig.add_trace(go.Scatter(
      x=marker_x,
      y=marker_y,
      mode='markers',
      marker=dict(color='royalblue', size=6),
      name="Discreto",  # Oculta na legenda
      hovertemplate="<b>Sinal Discreto</b><br>Amostra: %{x:.0f}<br>Valor: %{y:.5f}<extra></extra>",
  ), row=3, col=1)

  for i in range(len(t)):
    lines_x += [i, i, None]
    lines_y += [0, sinal_processado[i], None]

  fig.add_trace(go.Scatter(
      x=lines_x,
      y=lines_y,
      mode='lines',
      line=dict(color='royalblue', width=2),
      name=None,
      hovertemplate="<extra></extra>",
      showlegend=False), row=3, col=1)
  marker_y.clear()
  marker_x.clear()
  lines_x.clear()
  lines_y.clear()

  for i in range(len(t)):
      if not np.isclose(sinal_processado[i], 0):
        marker_x.append(t[i])
        marker_y.append(sinal_processado[i])

  fig.add_trace(go.Scatter(
      x=marker_x,
      y=marker_y,
      mode='markers',
      marker=dict(color='royalblue', size=6),
      name="Discreto",  # Oculta na legenda
      showlegend=False,
      hovertemplate="<b>Sinal Discreto</b><br>Tempo: %{x:.3f}<br>Valor: %{y:.5f}<extra></extra>",
  ), row=2, col=1)

  for i in range(len(t)):
    lines_x += [t[i], t[i], None]
    lines_y += [0, sinal_processado[i], None]

  fig.add_trace(go.Scatter(
      x=lines_x,
      y=lines_y,
      mode='lines',
      line=dict(color='royalblue', width=2),
      name=None,
      hovertemplate="<extra></extra>",
      showlegend=False), row=2, col=1)

  fig.add_trace(go.Scatter(
      x=t,
      y=sinal_processado,
      mode='lines',
      line=dict(color='royalblue'),
      hovertemplate="<b>Sinal 'Contínuo'</b><br>Tempo: %{x:.3f} s<br>Valor: %{y:.5f}<extra></extra>",
      name='contínuo'), row=1, col=1)


  fig['layout']['xaxis2']['matches'] = 'x1'
  fig['layout']['yaxis2']['matches'] = 'y1'

  # Eixos
  fig.update_yaxes(title_text="Amplitude", row=1, col=1)
  fig.update_yaxes(title_text="Amplitude", row=2, col=1)
  fig.update_yaxes(title_text="Amplitude", row=3, col=1)
  fig.update_xaxes(title_text="Tempo (s)", row=1, col=1)
  fig.update_xaxes(title_text="Tempo (s)", row=2, col=1)
  fig.update_xaxes(title_text="Número de Amostras (N)", row=3, col=1)



  # Layout
  fig.update_layout(
      height=800,
      title="Sinal Senoidal (50 Hz) - Contínuo e Amostrado",
      hovermode='x unified'
  )

  fig.show()

```

# IMPLEMENTAÇÃO

```python
fs = 500 #frquência de amostragem
n = 1000 #número de pontos

sinal,tempo = gerador_da_funcao_senoide(amostrador(fs,n))

```

# Gráfico Gerado

```python
plot(sinal,tempo,fs)
```
